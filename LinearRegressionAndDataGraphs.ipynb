```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_cells = [11, 18, 14, 43, 12, 50, 58, 13]
y_time = [4, 8, 6, 19, 5, 19, 25, 7]
y_code = [27, 41, 33, 91, 29, 105, 121, 31]
```

```python
plt.plot([11, 12, 13, 14, 18, 43, 50, 58],
         [4, 5, 7, 6, 8, 19, 19, 25])
plt.xlabel('Кол-во ячеек')
plt.ylabel('Время, мин.')
plt.title('Зависимость времени, затраченного на подготовку пайплайна, от кол-ва ячеек')
plt.show()
```

```python
plt.plot([11, 12, 13, 14, 18, 43, 50, 58],
         [27, 29, 31, 33, 41, 91, 105, 121])

plt.xlabel('Кол-во ячеек')
plt.ylabel('Кол-во строк кода')
plt.title('Зависимость кол-ва добавленных строк кода от кол-ва ячеек')
plt.show()
```

```python
x = np.array(x_cells).reshape(-1, 1)
y = np.array(y_time).reshape(-1, 1)

reg = LinearRegression()
reg.fit(x, y)
```

```python
x_test = np.array([0, 15, 17, 20, 38, 44, 50, 70]).reshape(-1, 1)

prediction = reg.predict(x_test)
prediction
```

```python
plt.scatter(x, y, label='Данные')
plt.plot(x_test, prediction, label='Регрессия')

plt.title("Зависимость времени, затраченного на подготовку пайплайна, от кол-ва ячеек")
plt.xlabel('Кол-во ячеек')
plt.ylabel('Время, мин.')
plt.legend(loc='best')

plt.show()
```
